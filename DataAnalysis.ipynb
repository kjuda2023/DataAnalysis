```python
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
data=pd.read_excel("Monthly Trip Report.xlsx")
df=pd.DataFrame(data)
#Convert string to date
#d_parser= lambda x:datetime.strptime(x,"%Y-%m-%d")
#data2=pd.read_excel("Detailed Event Report.xlsx",parse_dates=['Start Date'], date_parser=d_parser)
data2=pd.read_excel("Detailed Event Report.xlsx", parse_dates=['Start Date'])
df1=pd.DataFrame(data2)
df1.info()
```

```python
pivot_all=pd.pivot_table(df, index=["Driver Name"],columns='Month',values=['Distance (km)','Days With Trips'],fill_value=0,aggfunc={'Distance (km)':np.sum,'Days With Trips':np.sum})
pivot_all.head(20)
```
