```python
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
data=pd.read_excel("Data/Monthly Trip Report.xlsx")
df=pd.DataFrame(data)
#Convert string to date
#d_parser= lambda x:datetime.strptime(x,"%Y-%m-%d")
#data2=pd.read_excel("Detailed Event Report.xlsx",parse_dates=['Start Date'], date_parser=d_parser)
data2=pd.read_excel("Data/Detailed Event Report.xlsx", parse_dates=['Start Date'])
data2['Start Date']=pd.to_datetime(data2['Start Date']).dt.month_name()
df1=pd.DataFrame(data2)
df1=df1.drop_duplicates()
df1=df1.fillna('')
#df1=df1.drop(columns=['Event Location','End Location','End Lat/Long','Event Value','Distance calculated from odo (km)'])
df1.info()
df1.describe()

```
